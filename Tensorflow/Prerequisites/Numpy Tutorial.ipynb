# The Dot Product

$$ a \cdot b = a^T b = \sum_{d=1}^D a_d b_d $$

```python
a = np.array([1, 2])
b = np.array([3, 4])
```

```python
dot = 0
for e, f in zip(a, b):
  dot += e * f
dot
```

```python
dot = 0
for i in range(len(a)):
  dot += a[i] * b[i]
dot
```

```python
a * b
```

```python
np.sum(a * b)
```

```python
(a * b).sum()
```

```python
np.dot(a, b)
```

```python
a.dot(b)
```

```python
a @ b
```

$$ a^T b = \left\lVert a \right\rVert \left\lVert b \right\rVert \cos \theta_{ab}$$

$$ \cos \theta_{ab} = \frac{a^T b}{\left\lVert a \right\rVert \left\lVert b \right\rVert} $$

$$ \left\lVert a \right\rVert = \sqrt{\sum_{d=1}^D a_d^2 } $$

```python
amag = np.sqrt((a * a).sum())
amag
```

```python
np.linalg.norm(a)
```

```python
cosangle = a.dot(b) / (np.linalg.norm(a) * np.linalg.norm(b))
cosangle
```

```python
angle = np.arccos(cosangle)
angle
```

# Speed Test

```python
## speed comparison ##
import numpy as np
from datetime import datetime

# note: you can also use %timeit

a = np.random.randn(100)
b = np.random.randn(100)
T = 100000

def slow_dot_product(a, b):
  result = 0
  for e, f in zip(a, b):
    result += e*f
  return result

t0 = datetime.now()
for t in range(T):
  slow_dot_product(a, b)
dt1 = datetime.now() - t0

t0 = datetime.now()
for t in range(T):
  a.dot(b)
dt2 = datetime.now() - t0

print("dt1 / dt2:", dt1.total_seconds() / dt2.total_seconds())
```

# Solving Linear Systems

$$ x_1 + x_2 = 2200\\
  1.5 x_1 + 4x_2 = 5050$$

```python
A = np.array([[1, 1], [1.5, 4]])
b = np.array([2200, 5050])
```

```python
np.linalg.solve(A, b)
```

```python
np.linalg.inv(A).dot(b)
```
