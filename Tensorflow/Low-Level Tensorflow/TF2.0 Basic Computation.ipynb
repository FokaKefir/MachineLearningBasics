```python
import tensorflow as tf
```

```python
a = tf.constant(3.0)
b = tf.constant(4.0)
c = tf.sqrt(a ** 2 + b ** 2)
print("c:", c)
print(f"c: {c}")
```

```python
c.numpy()
```

```python
type(c.numpy())
```

```python
a = tf.constant([1, 2, 3])
b = tf.constant([4, 5, 6])
c = tf.tensordot(a, b, axes=[0, 0])
print(f"c: {c}")
```

```python
a.numpy().dot(b.numpy())
```

```python
import numpy as np
A0 = np.random.randn(3, 3)
b0 = np.random.randn(3, 1)
c0 = A0.dot(b0)
print(f"c0: {c0}")
```

```python
A = tf.constant(A0)
b = tf.constant(b0)
c = tf.matmul(A, b)
print(f"c: {c}")
```

```python
A = tf.constant([[1, 2], [3, 4]])
b = tf.constant(1)
C = A + b
print(f"C: {C}")
```

```python
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[2, 3], [4, 5]])
C = A * B
print(f"C: {C}")
```
